```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
month_temp = pd.read_csv("Resources/ctempmonth.csv")
```

```python
month_temp
```

```python
month_temp.dtypes
```

```python
month_temp["Date"] = month_temp["Date"].astype(str)

month_temp['Year'] = month_temp["Date"].str[:4]

month_temp
```

```python
month_temp
```

```python
month_temp['Month'] = month_temp["Date"].str.strip().str[-2:]
```

```python
month_temp
```

```python
month_temp['Month'] = month_temp['Month'].replace('01', 'Jan', regex=True)
month_temp['Month'] = month_temp['Month'].replace('02', 'Feb', regex=True)
month_temp['Month'] = month_temp['Month'].replace('03', 'Mar', regex=True)
month_temp['Month'] = month_temp['Month'].replace('04', 'Apr', regex=True)
month_temp['Month'] = month_temp['Month'].replace('05', 'May', regex=True)
month_temp['Month'] = month_temp['Month'].replace('06', 'Jun', regex=True)
month_temp['Month'] = month_temp['Month'].replace('07', 'Jul', regex=True)
month_temp['Month'] = month_temp['Month'].replace('08', 'Aug', regex=True)
month_temp['Month'] = month_temp['Month'].replace('09', 'Sep', regex=True)
month_temp['Month'] = month_temp['Month'].replace('10', 'Oct', regex=True)
month_temp['Month'] = month_temp['Month'].replace('11', 'Noc', regex=True)
month_temp['Month'] = month_temp['Month'].replace('12', 'Dec', regex=True)
```

```python
month_temp
```

```python
del month_temp['Date']
month_temp
```

```python
month_temp.to_csv("Resources/monthly_temp.csv")
```
