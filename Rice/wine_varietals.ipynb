```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# read in csv's
white_wine = pd.read_csv("Resources/Varietals_CSV/White_Varietals.csv")
red_wine = pd.read_csv("Resources/Varietals_CSV/Red_Varietals.csv")
```

```python
wine_varietal = pd.merge(red_wine, white_wine, on=['Varietal', 'Type', '2011', '2012', '2013', '2014', '2015', '2016',
       '2017', '2018', '2019'], how='outer')
wine_varietal
```

```python
zero_wines = wine_varietal[(wine_varietal['2011'] == 0) & (wine_varietal['2012'] == 0) 
                           & (wine_varietal['2013'] == 0) & (wine_varietal['2014'] == 0)  
                           & (wine_varietal['2015'] == 0) & (wine_varietal['2016'] == 0) 
                           & (wine_varietal['2017'] == 0) & (wine_varietal['2018'] == 0) 
                           & (wine_varietal['2019'] == 0)]

zero_wines
```

```python
wine_varietal = wine_varietal.drop([7,8,31, 45, 46, 59]) 
wine_varietal

```
