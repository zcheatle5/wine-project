# Merge temperature and grape datasets

```python
# import dependancies
import pandas as pd
import numpy as np
```

```python
# read in temperature dataset
temp_df = pd.read_csv('ctemp.csv')
temp_df.head(5) 
```

```python
temp_df.to_csv('ctemp.csv')
```

```python
# read in temperature increase dataset
grape_df = pd.read_csv('cchange.csv')
grape_df.head(10)
```

```python
grape_df.to_csv('cchange.csv')
```

```python
# read in precipitation dataset

pcp_df = pd.read_csv('cpcp.csv')
pcp_df.head(10) 
```

```python
pcp_df.to_csv('cpcp.csv')
```

# merge temp and precip data and clean weather df

```python
 # Merge temp and precip using an outer join
weather_df = pd.merge(temp_df, pcp_df, on="Unnamed: 0", how="outer")
weather_df
```

```python
for column in weather_df:
    print (column)
```

```python
weather_df.head(10)
```

```python
weather_df = weather_df.rename(columns={"Unnamed: 0": "id", "County_x": "County",
                                        "Total Pcp": "Total_pcp",        
                                        "Year_y": "Year",
                                                })
weather_df
```

```python
weather_df.iloc[:, :3]
```

```python
weather_df.to_csv("weather.csv")
```

```python
# check for na values
weather_df.isna().sum()
```

```python
weather_df.fillna(value=0, inplace=True)
```
